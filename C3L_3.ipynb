```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker
from matplotlib import rcParams
pd.options.display.float_format = '{}'.format

import warnings
warnings.filterwarnings('ignore')
```

# Feb 19

```python
# Prepare file
df = pd.read_csv("http://dada2020.xyz/dada/Data.csv", dtype=str)
```

```python
df = df.iloc[:,0:11]
df[["Passive", "Active","Constructive","Interactive"]] = df[["Passive", "Active","Constructive","Interactive"]].apply(pd.to_numeric)
df = df.fillna(0)
df['Time'] = pd.to_datetime(df['Time stamp'])
df['ID'] = range(1538)
```

### Assign weight to PACI (1,2,3,4)

```python
# assign weight to PACI (1,2,3,4)
df['Active'] = df['Active']*2
df['Constructive'] = df['Constructive']*3
df['Interactive'] = df['Interactive']*4
```

```python
# split into different time zones
df_s1 = df[df['Time']<='2017-10-15'] #795
df_s2 = df[df['Time']>'2017-10-15'] #743
df_s1_t1 = df[df['Time']<='2017-09-20'] #412
df_s1_t2 = df[ (df['Time']>'2017-09-20') & (df['Time']<'2017-10-15') ] #383
df_s2_t1 = df[ (df['Time']>'2017-10-15')&(df['Time']<='2017-11-02') ] #437
df_s2_t2 = df[ (df['Time']>'2017-11-02') ] #306
```

# Density plot

### mean+ratio PACI by students

```python
#mean PACI for each student in semester1 t1
tmp = df_s1_t1
PACI_sum = tmp.groupby(['Student ID'])["Passive", "Active","Constructive","Interactive"].sum()
Student_count = tmp.groupby(['Student ID'])['ID'].count()
PACI_mean = PACI_sum.sum(axis=1)/Student_count
plt.hist(PACI_mean)
PACI_mean_stu_s1t1 = PACI_mean


# ratio of PACI for student in semester1 t1
a = PACI_sum.sum(axis=1)
b = PACI_sum/a[:,None]
fig, ax = plt.subplots(figsize=(5,4), dpi=100)
ax.set_xlim(0,1)
plt.hist(b['Active'], alpha=0.5, label='Active')
plt.hist(b['Passive'], alpha=0.5, label='Passive')
plt.hist(b['Constructive'], alpha=0.5, label='Constructive')
plt.hist(b['Interactive'], alpha=0.5, label='Interactive')
plt.legend()
```
